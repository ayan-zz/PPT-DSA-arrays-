```python
class ListNode:
    def __init__(self, val=0, prev=None, next=None):
        self.val = val
        self.prev = prev
        self.next = next

def reverseDoublyLinkedList(head):
    if not head:
        return head

    current = head

    while current:
        current.prev, current.next = current.next, current.prev
        if not current.prev:
            head = current
        current = current.prev

    return head


head = ListNode(1)
node2 = ListNode(2)
node3 = ListNode(3)
node4 = ListNode(4)
node5 = ListNode(5)

head.next = node2
node2.prev = head
node2.next = node3
node3.prev = node2
node3.next = node4
node4.prev = node3
node4.next = node5
node5.prev = node4

reversed_head = reverseDoublyLinkedList(head)

current = reversed_head
while current:
    print(current.val, end=" ")
    current = current.next
```

```python
class ListNode:
    def __init__(self, val=0, prev=None, next=None):
        self.val = val
        self.prev = prev
        self.next = next

def deleteNodeAtPosition(head, position):
    if not head:
        return head

    current = head

    for _ in range(position):
        if not current:
            return head
        current = current.next

    if current == head:
        head = head.next
        if head:
            head.prev = None
        return head

    if current.prev:
        current.prev.next = current.next
    if current.next:
        current.next.prev = current.prev

    return head



head = ListNode(1)
node2 = ListNode(2)
node3 = ListNode(3)
node4 = ListNode(4)
node5 = ListNode(5)

head.next = node2
node2.prev = head
node2.next = node3
node3.prev = node2
node3.next = node4
node4.prev = node3
node4.next = node5
node5.prev = node4


position = 2
new_head = deleteNodeAtPosition(head, position)


current = new_head
while current:
    print(current.val, end=" ")
    current = current.next
```
